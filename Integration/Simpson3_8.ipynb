```python
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defina a função que deseja integrar
def f(x):
    return 4*x**5 + 2*x**4 + 9*x**3 + 8*x**2 + 6*x + 6

# Defina os limites do intervalo
a = 0
b = 10
# Número de subintervalos
n = 1000
```

```python
# Implementação do Método de Simpson 3/8
def Simpson3_8(f, a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    y = f(x)
    integral = 0

    for i in range(0, n-2, 3):
        integral += (3*h / 8) * (y[i] + 3 * y[i+1] + 3 * y[i+2] + y[i+3])

    return integral
```

```python
# Medição do tempo de execução
start_time = time.time()

# Cálculo da integral usando o Método de Simpson 3/8
integral_result = Simpson3_8(f, a, b, n)

# Tempo de execução
execution_time = time.time() - start_time

# Impressão dos resultados
print("Resultado da integral:", integral_result)
print("Número de iterações:", n)
print("Tempo de execução:", execution_time, "segundos")

# Pontos para o gráfico
x = np.linspace(a, b, n+1)
y = f(x)

# Plotar o gráfico da função
plt.plot(x, y, 'b', label='Função')
plt.fill_between(x, y, color='blue', alpha=0.3, label='Área sob a curva')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Método de Simpson 3/8')
plt.legend()
plt.show()
```
