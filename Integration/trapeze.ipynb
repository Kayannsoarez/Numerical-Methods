```python
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defina a função que deseja integrar
def f(x):
    return 4*x**5 + 2*x**4 + 9*x**3 + 8*x**2 + 6*x + 6

# Defina os limites do intervalo
a = 0
b = 10
# Número de subintervalos
n = 1000
```

```python
# Implementação do Método do Trapézio
def Trapeze(f, a, b, n):
    h = (b - a) / float(n)
    s = 0.5 * (f(a) + f(b))
    for i in range(1, n):
        s = s + f(a + i * h)
    return h * s
```

```python
# Medição do tempo de execução
start_time = time.time()

# Cálculo da integral usando o Método do Trapézio
integral_result = Trapeze(f, a, b, n)

# Tempo de execução
execution_time = time.time() - start_time

# Impressão dos resultados
print("Resultado da integral:", integral_result)
print("Número de iterações:", n)
print("Tempo de execução:", execution_time, "segundos")

# Pontos para o gráfico
x = np.linspace(a, b, n+1)
y = f(x)

# Plotar o gráfico dos retângulos
plt.plot(x, y, 'b', label='Função')
plt.bar(x, y, width=(b-a)/n, alpha=0.3, align='edge', label='Retângulos')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Método do Trapézio')
plt.legend()
plt.show()
```
