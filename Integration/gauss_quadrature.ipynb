```python
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defina a função que deseja integrar
def f(x):
    return 4*x**5 + 2*x**4 + 9*x**3 + 8*x**2 + 6*x + 6

# Definição dos limites de integração
a = -1
b = 1

# Pontos e pesos da Quadratura de Gauss
gauss_points = [-0.5773502692, 0.5773502692]
gauss_weights = [1.0, 1.0]
```

```python
# Implementação do Método da Quadratura de Gauss
def gauss_quadrature(f, a, b):
    integral = 0.0
    for i in range(len(gauss_points)):
        x = 0.5 * (b - a) * gauss_points[i] + 0.5 * (b + a)
        integral += 0.5 * (b - a) * gauss_weights[i] * f(x)
    return integral
```

```python
# Medição do tempo de execução
start_time = time.time()

# Cálculo da integral usando o Método da Quadratura de Gauss
integral_result = gauss_quadrature(f, a, b)

# Tempo de execução
execution_time = time.time() - start_time

# Impressão dos resultados
print("Resultado da integral:", integral_result)
print("Número de iterações:", len(gauss_points))
print("Tempo de execução:", execution_time, "segundos")

# Pontos para o gráfico
x = np.linspace(a, b, 100)
y = f(x)

# Plotar o gráfico da função
plt.plot(x, y, 'b', label='Função')
plt.fill_between(x, y, color='blue', alpha=0.3, label='Área sob a curva')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Método da Quadratura de Gauss')
plt.legend()
plt.show()
```
