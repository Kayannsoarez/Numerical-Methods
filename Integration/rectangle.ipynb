```python
import time
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Cria a função no qual a raiz deve ser calculada
def f(x):
    return 4*x**5 + 2*x**4 + 9*x**3 + 8*x**2 + 6*x + 6

# Define os limites do intervalo
a = 0
b = 10
# Número de retângulos (subintervalos)
n = 1000
```

```python
# Implementação do Método do Retângulo
def retangulo(f, a, b, n):

    # Largura de cada subintervalo
    h = (b - a) / n  
    soma = 0
    x = np.linspace(a, b, n+1)
    y = f(x)

    for i in range(n):
        # Área do retângulo
        area = y[i] * h
        soma += area

    return soma, x, y
```

```python
# Medição do tempo de execução
start = time.time()
integral_result, x, y = retangulo(f, a, b, n)
end = time.time()

execution_time = end - start

print("Resultado da integral:", integral_result)
print("Tempo de execução:", execution_time, "segundos")
print("Número de iterações:", n)

# Plotar o gráfico dos retângulos
plt.plot(x, y, 'b', label='Função')
plt.bar(x, y, width=(b-a)/n, alpha=0.3, align='edge', label='Retângulos')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Método do Retângulo')
plt.legend()
plt.show()
```
