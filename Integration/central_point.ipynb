```python
import math
import time
import matplotlib.pyplot as plt
```

```python
# Defina a função que deseja integrar
def f(x):
    return 4*x**5 + 2*x**4 + 9*x**3 + 8*x**2 + 6*x + 6

# Defina os limites do intervalo
a = 0
b = 10
# Número de subintervalos
n = 1000
```

```python
# Implemente o método do ponto central para integração numérica
def ponto_central(f, a, b, n):
    # Largura de cada subintervalo
    h = (b - a) / n  
    integral = 0

    for i in range(n):
        # Ponto médio do subintervalo
        x = a + (i + 0.5) * h
        integral += f(x) * h

    return integral
```

```python
# Calcule a integral usando o método do ponto central
start_time = time.time()
integral_result = ponto_central(f, a, b, n)
end_time = time.time()

execution_time = end_time - start_time
iterations = n

print("Resultado da integral:", integral_result)
print("Tempo de execução:", execution_time, "segundos")
print("Número de iterações:", iterations)

# Gere o gráfico dos retângulos
x = []
y = []
h = (b - a) / n
for i in range(n):
    # Ponto médio do subintervalo
    xi = a + (i + 0.5) * h
    yi = f(xi)
    x.append(xi)
    y.append(yi)

# Plota o gráfico dos retângulos
plt.plot(x, y, 'b', label='Função')
plt.bar(x, y, width=(b-a)/n, alpha=0.3, align='edge', label='Retângulos')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Método do Ponto Central')
plt.legend()
plt.show()
```
